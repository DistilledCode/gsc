## Zero Skewness $\not\implies$ Symmetric Distribution
--- 

Zero skewness only tells that the deviation on both end of mean are balanced.

https://papers.ssrn.com/sol3/papers.cfm?abstract_id=2531847

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from random import random
from statistics import NormalDist
import pandas as pd


# Set the seed for reproducibility
np.random.seed(42)

# Generate the data for the three cases
n = 10_000_000  # Sample size
d1 = NormalDist(mu=1, sigma=np.sqrt(2)).samples(2*(n//3))
d2 = NormalDist(mu=-2, sigma=1).samples(n//3)

d1.extend(d2)
l = d1
```

```python
l = pd.Series(l)
```

```python
l.mean(), l.skew(), l.median()
```

```python
l.min(), l.max()
```

```python
sns.kdeplot(l)
```
